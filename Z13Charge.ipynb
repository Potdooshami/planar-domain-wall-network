```python
class Z13:
    def __init__(self, value):
        if not isinstance(value, int):
            raise TypeError("Value must be an integer")
        self.value = value % 13

    def __add__(self, other):
        if isinstance(other, Z13):
            return Z13(self.value + other.value)
        return Z13(self.value + other)

    def __sub__(self, other):
        if isinstance(other, Z13):
            return Z13(self.value - other.value)
        return Z13(self.value - other)

    #def __mul__(self, other):
        #if isinstance(other, Z13):
         #   return Z13(self.value * other.value)
        #return Z13(self.value * other)

    def __eq__(self, other):
        if isinstance(other, Z13):
            return self.value == other.value
        return False

    def __repr__(self):
        return f"Z13({self.value})"

    def inverse(self):
        for i in range(13):
            if (self.value * i) % 13 == 1:
                return Z13(i)
        raise ValueError(f"No multiplicative inverse for {self.value} in Z13")
    def val(self):
        return self.value

# Example usage:
elements = [Z13(i) for i in range(13)]

# Printing the elements of Z13
for element in elements:
    print(element)

# Example operations
a = Z13(5)
b = Z13(8)
print(f"{a} + {b} = {a + b}")
print(f"{a} - {b} = {a - b}")
#print(f"{a} * {b} = {a * b}")

# Finding the multiplicative inverse of an element
print(f"The multiplicative inverse of {a} is {a.inverse()}")
```

Vortex case

```python
foo = []
for ind in range(1):
  for jnd in range(ind+1,13):
    for knd in range(jnd+1,13):
      foo.append([ind,jnd,knd])
```

```python
import numpy as np
vortBF = np.array(foo)
type(vortBF)
vortBF.shape
```

```python
dws = np.mod(vortBF[:,[1,2,0]] - vortBF,13)
com3 = np.unique(np.sort(dws),axis=0)
com3
```

```python
ishelic = []
for ind in range(len(com3)):
  ishelic.append(len(np.unique(com3[ind,:]))==2)
ishelic = np.array(ishelic)
ishelic
```

```python
hel0 = com3[ishelic,:]
hel1 = com3[np.logical_not(ishelic),:]
heln = hel1[:,::-1]

vorts = np.concatenate((hel1,hel0,heln),axis=0)
foo1 = np.ones((len(hel1),1))
foo0 = np.zeros((len(hel0),1))
hels = np.concatenate((foo1,foo0,-foo1),axis=0)
import pandas as pd
vortAll = pd.DataFrame(vorts,columns=['w1','w2','w3'])
vortAll['helicity'] = hels
vortAll
```

```python
vortAll.iloc[:,:-1].values
```

```python
antVort = np.mod(-vortAll.iloc[:,:-1].values,13)
antVort = np.concatenate((antVort,vortAll.iloc[:,-1].values.reshape(-1,1)),axis=1)
antVort = pd.DataFrame(antVort,columns = vortAll.columns)
vortAll = pd.concat((vortAll,antVort),axis=0)
vortAll.reset_index(drop=True)
```
